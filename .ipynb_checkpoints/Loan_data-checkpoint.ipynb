- Convert Listing Creation Date and Closed Date to Date Time object
- Group Key, TotalProsperLoans, TotalProsperPaymentsBilled, OnTimeProsperPayments, ProsperPaymentsLessThanOneMonthLate, ProsperPaymentsOneMonthPlusLate, ProsperPrincipalBorrowed, ProsperPrincipalOutstanding, ScorexChangeAtTimeOfListing, LoanFirstDefaultCycleNumber, and CreditGrade all have missing data. I will delete them all.
- 

```python
# remove mostly null columns

remove_string = "GroupKey, TotalProsperLoans, TotalProsperPaymentsBilled, OnTimeProsperPayments, ProsperPaymentsLessThanOneMonthLate, ProsperPaymentsOneMonthPlusLate, ProsperPrincipalBorrowed, ProsperPrincipalOutstanding, ScorexChangeAtTimeOfListing, LoanFirstDefaultedCycleNumber, CreditGrade"
```

```python
remove_columns = []

for string in remove_string.split(","):
    remove_columns.append(f"{string}".lstrip())
```

```python
df = df.drop(remove_columns, axis=1)
```

```python
df["ListingCreationDate"] = pd.to_datetime(df["ListingCreationDate"])
```

```python
df["CreationDate"] = df["ListingCreationDate"].dt.date
```

```python
df["CreationHour"] = df["ListingCreationDate"].dt.hour
```

```python
df["ClosedDate"] = pd.to_datetime(df["ClosedDate"])
```

```python
df["ClosedHour"] = df["ClosedDate"].dt.hour
```

```python
df["ClosedDate"] = df["ClosedDate"].dt.date
```

```python
df.drop("ListingCreationDate", axis=1)
```

```python
states = df["BorrowerState"].unique()
```

```python
d = {}

for state in states:
    d[state] = df[df["BorrowerState"]==state]["LoanStatus"]
```
